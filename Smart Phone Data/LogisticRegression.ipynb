```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools
import random
from statistics import mean 
from sklearn import preprocessing
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, precision_score, recall_score,accuracy_score, confusion_matrix, f1_score


train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')


both_df = pd.concat([train_df, test_df], axis=0).reset_index(drop=True)

#print('Shape both:\t{}\n'.format(both_df.shape))

#c = list(range(1, 31))
#a=random.sample(c,k=5)
a=[1,2,3,4,5]

print('persons that are kept for testing:\t{}\n'.format(a))

train_df=both_df[(both_df.subject != a[0])&(both_df.subject != a[1])
         &(both_df.subject != a[2])&(both_df.subject != a[3])
         &(both_df.subject != a[4])] 

#print('Shape Train:\t{}'.format(train_df.shape))

test_df=both_df.loc[both_df['subject'].isin([a[0],a[1],a[2],a[3],a[4]])]

#print('Shape Test:\t{}\n'.format(test_df.shape))


trainData  = train_df.drop('Activity' , axis=1).values
trainLabel = train_df.Activity.values

testData  = test_df.drop('Activity' , axis=1).values
testLabel = test_df.Activity.values

encoder = preprocessing.LabelEncoder()

# encoding train labels 
encoder.fit(trainLabel)
trainLabelE = encoder.transform(trainLabel)
 
# encoding test labels 
encoder.fit(testLabel)
testLabelE = encoder.transform(testLabel)

classifier = LogisticRegression(random_state=0, multi_class='ovr')

classifier.fit(trainData, trainLabelE)
y_score = classifier.predict_proba(testData)
y_te_pred = classifier.predict(testData)
acc1 = accuracy_score(testLabelE, y_te_pred)
prec1 = precision_score(testLabelE, y_te_pred, average="macro")
rec1 = recall_score(testLabelE, y_te_pred, average="macro")
F1_score1= f1_score(testLabelE, y_te_pred, average="macro")  
cfs = confusion_matrix(testLabelE, y_te_pred)
print("Acc: %3.5f, P: %3.5f, R: %3.5f, f1_score: %3.5f" % (acc1, prec1, rec1,F1_score1))




def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    
    #This function prints and plots the confusion matrix.
    #Normalization can be applied by setting `normalize=True`.
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    #else:
        #print('Confusion matrix, without normalization')

    #print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')


plt.figure()
class_names = encoder.classes_
plot_confusion_matrix(cfs, classes=class_names,
                      title='Logistic regression Confusion Matrix, without normalization')
```
